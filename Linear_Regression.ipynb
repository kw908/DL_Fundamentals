```python
import torch
import numpy as np
import matplotlib.pyplot as plt



def forward(x):
    return w * x

    # evaluating data points with Mean Square Error.
def loss(y_pred, y):
    return torch.mean((y_pred - y) ** 2)

# Creating a function f(X) with a slope of -5
X = torch.arange(-5, 5, 0.1).reshape(-1, 1)
Y_true = -5 * X

Y_train = Y_true + 0.4 * torch.randn(X.size())

# Plot the line in red with grids
# Plot and visualizing the data points in blue
plt.plot(X.numpy(), Y_train.numpy(), 'b+', label='Y')
plt.plot(X.numpy(), Y_true.numpy(), 'r', label='func')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid('True', color='y')
plt.show()

# initialize w
w = torch.tensor(-10.0, requires_grad=True)

step_size = 0.1
loss_list = []
iter = 10

for i in range (iter):
    # making predictions with forward pass
    Y_pred = forward(X)
    # calculating the loss between original and predicted data points
    l = loss(Y_pred, Y_true)
    # storing the calculated loss in a list
    loss_list.append(l.item())
    # backward pass for computing the gradients of the loss w.r.t to learnable parameters
    l.backward()
    # updateing the parameters after each iteration
    w.data = w.data - step_size * w.grad
    # zeroing gradients after each iteration
    w.grad.zero_()
    # priting the values for understanding
    # print('{},\t{},\t{}'.format(i, l.item(), w.item()))

# Plotting the loss after each iteration
plt.plot(loss_list, 'r')
plt.tight_layout()
plt.grid('True', color='y')
plt.xlabel("Epochs/Iterations")
plt.ylabel("Loss")
plt.show()
```
